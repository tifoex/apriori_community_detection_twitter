```python
unique_clusters = labeled_df['cluster'].unique()
unique_labels = labeled_df['label'].unique()
unique_glabels = labeled_df['gLabel'].unique()
```

```python
# Loop through each community and print the clusters for each node in the community
main_cluster={}
for i in unique_labels:
  new_df = labeled_df[labeled_df['label'] == i].copy()
  cluster_counts = {}
  for j in unique_clusters:
    count = len(new_df[new_df['cluster'] == j])
    cluster_counts[j] = count
    # Find the cluster with the maximum count
  max_count_cluster = max(cluster_counts, key=cluster_counts.get)
  print(f"For label {i}, the cluster with the maximum count is: {max_count_cluster}")
  main_cluster[i]=max_count_cluster
```

```python
for index, row in labeled_df.iterrows():
  labeled_df.at[index, 'class'] = main_cluster[row['label']]
labeled_df.head()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Assuming you have the true labels in the 'class' column and predictions in the 'cluster' column
# Replace 'labeled_df' with your actual DataFrame
true_labels = labeled_df['class']
predicted_labels = labeled_df['cluster']

# Step 2: Compute the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Step 3: Compute precision, recall, and F-measure
num_classes = len(cm)
precision = []
recall = []
f_measure = []

for i in range(num_classes):
    tp = cm[i, i]  # True positives
    fp = sum(cm[:, i]) - tp  # False positives
    fn = sum(cm[i, :]) - tp  # False negatives

    # Precision
    if tp + fp == 0:
        precision.append(0)
    else:
        precision.append(tp / (tp + fp))

    # Recall
    if tp + fn == 0:
        recall.append(0)
    else:
        recall.append(tp / (tp + fn))

    # F-measure
    if precision[i] + recall[i] == 0:
        f_measure.append(0)
    else:
        f_measure.append(2 * (precision[i] * recall[i]) / (precision[i] + recall[i]))

# You can store the results in a DataFrame if needed
results_df = pd.DataFrame({
    'Class': range(num_classes),
    'Precision': precision,
    'Recall': recall,
    'F-measure': f_measure
})

results_df
```
